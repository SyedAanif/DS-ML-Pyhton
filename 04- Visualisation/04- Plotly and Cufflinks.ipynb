# Plotly and Cufflinks
- `Plotly` is a library that allows us to create interactive plots that we can use in dashboards or websites (we can save them as html files or static images).
- `Cufflinks` allows us to connect `Pandas` to `Plotly`.

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode,plot,iplot
```

```python
print(__version__)
```

```python
import cufflinks as cf
```

```python
# connect JS to notebook
init_notebook_mode(connected=True)
```

```python
# use cufflinks offline
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
df2.head()
```

## Various Plots

* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble

```python
df.iplot()
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers')
```

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df3.head()
```

```python
df3.iplot(kind='surface',colorscale='rdylbu') # rd- red, yl- yellow, bu- blue
```

```python
df.iplot(kind='hist')
```

```python
df[['A','D']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble',x='A',y='B',size='D')
```

```python
df.scatter_matrix()
```
