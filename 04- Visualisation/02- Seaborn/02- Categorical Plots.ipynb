# Categorical Data Plots

There are a few main plot types for this:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.info()
```

```python
# plots categorical vs mean(default) of continuous, with a confidence interval
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
# estimator=np.std for standard deviation
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='tip',data=tips)
```

```python
sns.boxplot(x='smoker',y='tip',data=tips,hue='time')
```

```python
sns.violinplot(x='day',y='tip',data=tips)
```

```python
sns.violinplot(x='smoker',y='tip',data=tips,hue='time')
```

```python
sns.violinplot(x='smoker',y='tip',data=tips,hue='time',split=True)
```

```python
sns.violinplot(y='smoker',x='tip',data=tips,hue='time',split=True,orient='h')
```

```python
sns.stripplot(x='smoker',y='total_bill',data=tips)
```

```python
sns.stripplot(x='smoker',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='smoker',y='total_bill',data=tips,hue='sex')
```

```python
#dodge and split are same
sns.stripplot(x='smoker',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.swarmplot(x='smoker',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='smoker',y='total_bill',data=tips,hue='sex')
```

```python
sns.swarmplot(x='smoker',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.violinplot(x='smoker',y='total_bill',data=tips,color='black')
sns.swarmplot(x='smoker',y='total_bill',data=tips)
```

```python
#generalised plot - catplot or factorplot
sns.catplot(x='smoker',y='total_bill',data=tips,kind='violin')
```
